```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
source = 'crime_trend_data.csv'

crime_df = pd.read_csv(source)
crime_df
```

```python
month = crime_df['Month']
savage_cases = crime_df['Savage_Cases']
```

```python
plt.figure(figsize=(10, 4))

plt.plot(month, savage_cases)

plt.title("Crime Trend Data")
plt.xlabel('Month',size=10)
plt.ylabel('Savage_Cases',size=10)

plt.grid(linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
source = 'biometrics_data.csv'

biometrics_df = pd.read_csv(source)
biometrics_df
```

```python
import seaborn as sns

plt.figure(figsize=(20, 8))

# สร้าง Scatter Plot
plt.scatter(biometrics_df['Body_Mass_kg'], biometrics_df['Food_Intake_kg'], color='orange', edgecolors='black', s=100)

sns.scatterplot(data=biometrics_df, x='Body_Mass_kg', y='Food_Intake_kg',
                hue='Type', s=150, palette=['red', 'blue'], edgecolor='black')

# ใส่ชื่อหัวข้อและชื่อแกน [cite: 79]
plt.title("Zootopia Biometrics: Body Mass vs Food Intake", fontsize=14)
plt.xlabel("Body Mass (kg)", fontsize=12)
plt.ylabel("Food Intake (kg/day)", fontsize=12)

# ใส่ชื่อสายพันธุ์กำกับที่จุด (เพื่อความชัดเจน)
for i, txt in enumerate(biometrics_df['Species']):
    plt.annotate(txt, (biometrics_df['Body_Mass_kg'][i], biometrics_df['Food_Intake_kg'][i]),
                 xytext=(5,5), textcoords='offset points')

plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
